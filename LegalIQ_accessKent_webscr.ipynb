```python
# test trail with these names before testing it on 3 million records.
names = ['neo','asd','leo']
```

```python
my_list=[]
for i in names:   # for each name
    try:
        # fill in the lastname blank
        browser.find_element_by_id('lastName').send_keys(i) 

        #Also fill in the date blanks
        browser.find_element_by_id('startYear').send_keys('2016')
        browser.find_element_by_id('endYear').send_keys('2017')


        #browser.find_element_by_xpath('/html/body/div[2]/div/div[1]/div/div/div/form/fieldset/input[1]').click()
        browser.find_element_by_name('submit').click()

        page =browser.find_element_by_xpath('/html/body/div[2]/div/div[1]/div/form/div/div[2]').text.split()  # for pg.No
        pg = int(math.ceil(int(page[3])/10))

        for r in range(pg):   # iterating through page numbers
            x = browser.find_element_by_xpath('/html/body/div[2]/div/div[1]/div/form/table')

            for i in range(1,len(x.find_elements_by_tag_name('tr'))):
                for j in range(3,12):

                    get_attr = browser.find_element_by_xpath('/html/body/div[2]/div/div[1]/div/form/table/tbody/tr[{}]/td[{}]'.format(i,j)).text
                    my_list.append(get_attr)
            browser.find_element_by_xpath('/html/body/div[2]/div/div[1]/div/form/div/div[3]/input[1]').click()

        browser.find_element_by_xpath('/html/body/div[2]/div/div[1]/div/form/ul/li[1]/a').click()
        print('done')
    
    except:
        browser.find_element_by_xpath('/html/body/div[2]/div/div[1]/div/form/ul/li[1]/a').click()
```
